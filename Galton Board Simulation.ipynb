```python
from random import randint
import matplotlib.pyplot as plt

def galton(h, b):
    containers = [0] * (h + 1)
    for _ in range(b):
        pos = 0
        for _ in range(h):
            turn = randint(0, 1)
            if turn == 1:
                pos += 1
        containers[pos] += 1
    return containers

def plot_graph(dataset):
    plt.bar(range(len(dataset)), dataset)
    plt.xlabel('Container Number')
    plt.ylabel('Balls in Container')
    plt.title('Distribution of Balls in Galton Board')
    plt.show()

# Prompt the user for input instead of using command-line arguments
try:
    h = int(input("Enter the height of the Galton board: "))
    b = int(input("Enter the number of balls: "))
    containers = galton(h, b)
    plot_graph(containers)
except ValueError as e:
    print(f"Invalid input: {e}")
```
